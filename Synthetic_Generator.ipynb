```python
import numpy as np
import random
import pandas as pd
#from BlackBoxAuditing.data import load_from_file

iq = np.array(np.random.randn(5000)) * 20 + 100
gender = [random.choice(["man", "woman"]) for i in range(5000)]
sat = [i * 10 + (np.random.randint(0,100) if g == "man" else np.random.randint(0,80)) for (i, g) in zip(iq, gender)]
admit = ["True" if (s > np.random.randint(750,1500))  else "False" for s in sat]

df = pd.DataFrame(
    {"admit": admit,
     "gender": gender,
     "iq": iq,
     "sat": sat})
df.to_csv("synthetic.csv", 
          index=False, 
          columns=['gender', 'admit', 'iq', 'sat']) # Make sure this order matches the order you're loading below
```
